```python
''' 
Nella serie di post che inizio in questa sezione, intendo schematizzare le fasi principali del processo di 
Natural language processing. In seguito differenzierò i processi in base alle varie applicazioni. 
Per il momento mi soffermo sulla fase di Pre-processing in cui modelliamo i nostri dati di input rendendoli 
digeribili alle reti neurali.

La prima cosa che vado a implementare è la Tokenization, ovvero la fase in cui il nostro input viene ridotto in unità elementari, in questo caso
le singole parole. Mi avvalgo delle risorse offerte da Python e in particolare della libreria NLTK molto utile in questo senso.
'''

import nltk
frase = "George is a really bad guy, but his girlfriend, Julie, is very pretty and I confess that I'd like to go out with her"

tokens = nltk.WhitespaceTokenizer().tokenize(frase)
print(tokens)
```

```python
tokens = nltk.WordPunctTokenizer().tokenize(frase)
print(tokens)
```

```python
tokens = nltk.TreebankWordTokenizer().tokenize(frase)
print(tokens)
```

```python
stemmer = nltk.stem.PorterStemmer()
' '.join(stemmer.stem(token) for token in tokens)
```

```python
stemmer = nltk.stem.WordNetLemmatizer()
' '.join(stemmer.lemmatize(token) for token in tokens)
```

```python
' '.join(token.lower() for token in tokens)
```
